```python
#PROG2
# car loan payment
def car_loan(P,Y,R):
    """Computes monthly car payment"""
    r= R / 100 /12
    n= Y *12
    monthly_payment = (P*r*(1+r)**n) / ((1+r) ** n-1)
    return monthly_payment

principal_amount = int(input("Enter the principal amount:"))
years = int(input("Enter the total number years to pay off:"))
roi = int(input("Enter the rate of interest:"))

monthly_payment = car_loan(principal_amount,years,roi)
print(f"Monthly car loan payment: {monthly_payment}")
```

```python
# PROG 3

def factorial(num):
    if num == 0:
        return 1
    else:
        result = num * factorial(num-1)
    
    return result

number = int(input("Enter Number:"))
result=factorial(number)

print(f"The factorial of the number {number} is {result}")
```
