```python
#PROG2
import sys
import timeit

tup = (1,2,3,4,5,6,7,8,9)
lst = [1,2,3,4,5,6,7,8,9]

tip_size=sys.getsizeof(tup)
lst_size=sys.getsizeof(lst)

print("Size of Tuple is",tip_size,"bytes")
print("Size of list is",lst_size,"bytes")

tup_time = timeit.timeit(stmt=lambda: tup,number=1000000)
lst_time = timeit.timeit(stmt=lambda: lst,number=1000000)

print("Creation time of tuple is",tup_time,"seconds")
print("Creation time of list is",lst_time,"seconds")
```
