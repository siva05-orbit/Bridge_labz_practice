```python
#Lambda
#PROG3

num = [1,2,3,4,5]
print("Original list is",num)
double = lambda x:x*2
print("Doubled list is",list(map(double,num)))
```

```python
#PROG4
#Remove element
def remove_element(color_list:list,indexs_to_remove:list):
    revised_list = []

    for i in range(len(color_list)):
        if i not in indexs_to_remove:
            revised_list.append(color_list[i])
            
    return revised_list

color_list = ["Red","Green","Yellow","White","Black","Orange","Purple","Pink"]
index = [2,3,5]

ans=remove_element(color_list=color_list,indexs_to_remove=index)
print("List before removing the elements:",color_list)
print("\nList after removing the elements:",ans)
```

```python
#Matrix Transpose
#PROG5

transposed_matrix=[[0,0,0],[0,0,0],[0,0,0]]

def matrix_transpose(matrix) :
    ROWS,COLS = len(matrix),len(matrix[0])
    for row in range(ROWS):
        for col in range(COLS):
            transposed_matrix[row][col] = matrix[col][row]
    return transposed_matrix

matrix = [[1,2,3],[4,5,6],[7,8,9]]
ans = matrix_transpose(matrix=matrix)

print("Before Transpose:")
for i in matrix:
    print(i)
print("After Transpose:")
for i in ans:
    
    print(i)
```
