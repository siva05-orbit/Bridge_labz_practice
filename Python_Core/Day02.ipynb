```python
#PROG 4
year=int(input("Enter Birth Year:"))

def find_generation(birth_year:int):
    
    print(f"The Birth Year is {birth_year}",
          f"Is Boomer: {birth_year >=1946 and birth_year < 1963}",
          f"Is GenX: {birth_year >=1963 and birth_year < 1980}",
          f"Is Millennial: {birth_year >=1980 and birth_year<1996}",
          f"Is GenZ: {birth_year >= 1996 and birth_year < 2012}",
          f"Is GenAlpha: {birth_year >= 2012 and birth_year < 2025}", sep = '\n')   
find_generation(year)
```

```python
#PROG 5 - time conversion
#output format : HH:MM:SS

total_seconds = int(input("Enter total seconds"))  

hours = total_seconds // 3600
minutes = (total_seconds % 3600) // 60
seconds = total_seconds % 60   

print(f"Time Duration of {total_seconds} seconds in HH:MM:SS format is \n"+f"{hours:02d}:{minutes:02d}:{seconds:02d}")
```

```python
# PROG 6 BILL GENERATION

apple_gst = 0.12
orange_gst = 0.05

buyer_name = input("Enter Buyer Name:")
apple_price = int(input("Enter price of apple per kg:"))
orange_price = int(input("Enter price of orange per kg:"))
apple_quant = float(input("Enter the quantity of apple in kg:"))
orange_quant = float(input("Enter the quantity og orange in kg:"))


apple_total = apple_price * apple_quant
orange_total = orange_price * orange_quant

apple_gst_total = apple_total * apple_gst
orange_gst_total = orange_total * orange_gst 

apple_billing = apple_total + apple_gst_total
orange_billing = orange_total + orange_gst_total

final_total = apple_billing + orange_billing
final_round_total = round(final_total)


#PRINTING THE BILL

print(f"Buyer's Name: {buyer_name}")
print("-"*84)
print(f"| {'Item':^12} | {'Price/kg':^10} | {'Quantity(kg)':^12} | {'Price':^10} | {'GST':10} | {'Total w/GST':^10} |")
print(f"-"*84)
print(f"| {'Apple':^12} | {apple_price:^10} | {apple_quant:^12} | {apple_total:^10} | {apple_gst_total:^10} | {apple_billing:^10} |")
print(f"| {'Orange':^12} | {orange_price:^10} | {orange_quant:^12} | {orange_total:^10} | {orange_gst_total:^10} | {orange_billing:^10} |")
print("-" * 84)
print(f"| {'Total':^12} | {'':^55} {final_total}  |")
print(f"| {'Round Total':^12} | {'':^55} {final_round_total}    |")
print("-"*84)


```

```python
#PROG 6
def school_report(total_students):

    for i in range(total_students):
        student_name=input("Enter Student Name:")
        school = 'DELHI PUBLIC SCHOOL'
        student_class = 'Class XII'
        chemistry_marks = int(input("Enter Chemistry Marks:"))
        physics_marks = int(input("Enter Physics Marks:"))
        maths_marks = int(input("Enter Maths Marks:"))

        total_marks = 100

        chemistry_p = round((chemistry_marks / total_marks) * 100,2)
        physics_p = round((chemistry_marks/total_marks)*100,2)
        maths_p = round((maths_marks/total_marks)*100,2)
        obtained_total=chemistry_marks+physics_marks+maths_marks

        avg_percentage = round((chemistry_p+physics_p+maths_p) / 3,2)
        #PRINTING REPORT:

        print(f"{school} - {student_class} - {student_name}")
        print(f"-"*60)
        print(f"| {'Subject':^10} | {'Total Marks':^12} | {'Obtained Marks':^12} | {'Percentage':^10} |")
        print(f"| {'Physics':^10} | {total_marks:^12} | {physics_marks:^12} | {physics_p:^10} |")
        print(f"| {'Chemistry':^10} | {total_marks:^12} | {chemistry_marks:^12} | {chemistry_p:^10} |")
        print(f"| {'Maths':^10} | {total_marks:^12} | {maths_marks:^12} | {maths_p:^10} |")
        print(f"-"*60)
        print(f"| {'Total':^10} | {'300':^12} | {obtained_total:^12} | {avg_percentage:^10} |")
        print("-"*60)

school_report(3)

```
