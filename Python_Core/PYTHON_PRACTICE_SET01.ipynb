```python
#PROBLEM 5
def salary_cal(salary:int,late_days:int,absent_days:int) -> int:
    final_salary = salary

    if late_days>10:
        late_penalty = final_salary/100 * 10
        final_salary = final_salary - late_penalty
    elif late_days>5:
        late_penalty = final_salary/100 * 5
        final_salary = final_salary - late_penalty
    
    if absent_days>2:
        absent_penalty = final_salary/100 * 5
        final_salary = final_salary - absent_penalty
    
    return final_salary

salary = int(input("Enter your Salary:"))
late_days = int(input("Enter total number of late days:"))
absent_days = int(input("Enter total number of absent days:"))

salary_cal(salary,late_days,absent_days)
```

```python
#PROBLEM 6
def prime_count(num1,num2):
    res=[]
    for i in range(num1,num2+1):
        is_prime=True
        for j in range(2,int(i**0.5+1)):
            if i%j==0:
                is_prime=False
                break
        if is_prime==True:
            res.append(i)
    
    return len(res)

a=int(input("Enter number 1:"))
b=int(input("Enter number 2:"))

prime_count(a,b)
```

```python
#PROBLEM 7
def discount(amount:int)->int:
    price = amount
    if amount >= 5000:
        dis = amount/100 *20
        price = price - dis
    elif amount >= 3000:
        dis = amount/100 * 10
        price = price - dis
    elif amount >= 1000:
        dis =amount/100 * 5
        price = price - dis
    return price

discount(3200)
```

```python
#PROBLEM 8
def binary_to_decimal(num):
    str_num=str(num)
    str_num = str_num[::-1]
    power=0
    res=0
    for digit in str_num:
        res=res+int(digit) * (2**power)
        power+=1 

    return res

binary_to_decimal(101101)
```

```python
#PROBLEM 11

def counter_a(num):
    count=0
    if num % 2 !=0:
        print("0")
    
    is_even=True

    while is_even:
        ans = num //2
        count+=1
        if ans % 2!=0:
            is_even=False
    
    return count
    
    
counter_a(40)
```

```python
#PROBLEM 13

def ticket_fare(distance,age):
    fare = distance * 2

    if age > 60:
        print("Scenior Citizen 30% Discount Applied")
        discount = fare/100 * 30
        fare = fare - discount
    elif age<12:
        print("Children Discount 50% applied")
        fare = fare * 0.5
    return int(fare)

ticket_fare(200,16)
```

```python
#Problem 15


def smart_lock(password):
    unlocked=False
    for i in range(1,4):
        a=int(input(f"Attempt {i}"))
        if str(a)==str(password):
            unlocked = True
            break
    if unlocked:
        print("ACCESS GRANTED")
    else:
        print("LOCKED")

a=int(input("Enter the password:"))
smart_lock(a)

```

```python
#PROBLEM 17
def armstrong(num):
    temp = num
    sum_c=0
    while temp>0:
        digit=temp%10
        temp = temp//10
        sum_c=sum_c + digit ** 3
    if sum_c==num:
        print("YES")
    else:
        print("NO")

armstrong(153)
```

```python
#PROBLEM 19

def palindrome(num):
    str_num = str(num)

    reverse = str_num[::-1]
    if str_num == reverse:
        return "PALINDROME"
    else:
        return "NOT PALINDROME"
palindrome(12121)
```

```python
#Problem 20

def countdown(num):
    temp=num
    while num>10:
        digit=temp%10
        temp=temp//10
        num=num-digit
    return num

countdown(987)
```
