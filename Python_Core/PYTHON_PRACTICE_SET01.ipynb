```python
#PROBLEM 2

def electricity(units):
    total_bill=0
    if units <= 100:
        total_bill=units*3
    elif units <=200:
        total_bill=100*3 + (units-100)*5
    else:
        total_bill=100*3 + 100*5 + (units-200)*8
    
    if units > 300:
        surcharge = total_bill * 0.10
        total_bill = total_bill+surcharge
    
    return int(total_bill)

electricity(350)
```

```python
#PROBLEM 5
def salary_cal(salary:int,late_days:int,absent_days:int) -> int:
    final_salary = salary

    if late_days>10:
        late_penalty = final_salary/100 * 10
        final_salary = final_salary - late_penalty
    elif late_days>5:
        late_penalty = final_salary/100 * 5
        final_salary = final_salary - late_penalty
    
    if absent_days>2:
        absent_penalty = final_salary/100 * 5
        final_salary = final_salary - absent_penalty
    
    return final_salary

salary = int(input("Enter your Salary:"))
late_days = int(input("Enter total number of late days:"))
absent_days = int(input("Enter total number of absent days:"))

salary_cal(salary,late_days,absent_days)
```

```python
#PROBLEM 6
def prime_count(num1,num2):
    res=[]
    for i in range(num1,num2+1):
        is_prime=True
        for j in range(2,int(i**0.5+1)):
            if i%j==0:
                is_prime=False
                break
        if is_prime==True:
            res.append(i)
    
    return len(res)

a=int(input("Enter number 1:"))
b=int(input("Enter number 2:"))

prime_count(a,b)
```

```python
#PROBLEM 7
def discount(amount:int)->int:
    price = amount
    if amount >= 5000:
        dis = amount/100 *20
        price = price - dis
    elif amount >= 3000:
        dis = amount/100 * 10
        price = price - dis
    elif amount >= 1000:
        dis =amount/100 * 5
        price = price - dis
    return price

discount(3200)
```

```python
#PROBLEM 11

def counter_a(num):
    count=0
    if num % 2 !=0:
        print("0")
    
    is_even=True

    while is_even:
        ans = num //2
        count+=1
        if ans % 2!=0:
            is_even=False
    
    return count
    
    
counter_a(40)
```

```python
#PROBLEM 12
def vowel_count(string):
    count=0
    

    for i in string.lower():
        if i in 'aeiou':
            count+=1
    
    return count

vowel_count("I Love Python")
```

```python
#PROBLEM 13

def ticket_fare(distance,age):
    fare = distance * 2

    if age > 60:
        print("Scenior Citizen 30% Discount Applied")
        discount = fare/100 * 30
        fare = fare - discount
    elif age<12:
        print("Children Discount 50% applied")
        fare = fare * 0.5
    return int(fare)

ticket_fare(200,16)
```

```python
#PROBLEM 14

def pattern_val(digits):
    b=str(digits)
    higher = False
    for i in range(len(b)):
        if int(b[i])>int(b[i+1]):
            higher = True
            break

    if higher:
        print("NO")
    else:
        print("YES")
a=1234
pattern_val(a)
```
