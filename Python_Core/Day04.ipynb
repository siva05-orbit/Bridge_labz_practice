```python
# EMI Calculator
def calculate_emi(principle, roi, time):
    monthly_rate = roi / (12 * 100)
    months = time * 12
    emi = principle * monthly_rate * (1 + monthly_rate) ** months / ((1 + monthly_rate) ** months - 1)
    return round(emi, 2)


def compute_emi_details(principle, roi, time):
    emi = calculate_emi(principle, roi, time)
    emi_details = []
    balance = principle
    months = time * 12

    for month in range(1, months + 1):
        interest = round(balance * roi / (12 * 100), 2)
        principal_paid = round(emi - interest, 2)
        balance = round(balance - principal_paid, 2)

        if balance < 0:
            principal_paid += balance
            balance = 0

        emi_details.append({
            "Month": month,
            "EMI": emi,
            "Interest": interest,
            "Principal": principal_paid,
            "Balance": balance
        })

        if balance == 0:
            break

    return emi_details


def display_emi_details(emi_details):
    print(f"{'-' * 80}")
    print(f"{'Month':^10}{'EMI':^15}{'Interest':^15}{'Principal':^15}{'Balance':^15}")
    print(f"{'-' * 80}")

    total_interest = 0
    total_principal = 0

    for row in emi_details:
        print(f"{row['Month']:^10}{row['EMI']:^15.2f}{row['Interest']:^15.2f}{row['Principal']:^15.2f}{row['Balance']:^15.2f}")
        total_interest += row['Interest']
        total_principal += row['Principal']

    print(f"{'-' * 80}")
    print(f"{'TOTAL':^10}{'':^15}{total_interest:^15.2f}{total_principal:^15.2f}{'':^15}")
    print(f"{'-' * 80}")


# MAIN PROGRAM

principle_entered = input("Enter the Principle Amount: ")
roi_entered = input("Enter the Rate of Interest (annual %): ")
time_entered = input("Enter the Time (in years): ")

principle = int(principle_entered) if principle_entered.isdigit() else None
roi = float(roi_entered) if roi_entered.replace('.', '', 1).isdigit() else None
time = int(time_entered) if time_entered.isdigit() else None

if principle and roi and time:
    emi_details = compute_emi_details(principle, roi, time)
    display_emi_details(emi_details)
else:
    print("Invalid input. Please enter numeric values.")
```
